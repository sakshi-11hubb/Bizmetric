```python
# usnig oops:
class CourseFeeCalculator:

    def __init__(self, subject, analytics, hostel, food_months, transport):
        self.available_courses = ["HR", "Finance", "Marketing", "DS"]
        self.base_fee = 200000
        self.subject = subject
        self.analytics = analytics.upper()
        self.hostel = hostel.upper()
        self.food_months = food_months
        self.transport = transport.lower()
        self.total_cost = 0

    def calculate_fee(self):

        # Course fee
        if self.subject in self.available_courses:
            self.total_cost += self.base_fee

            # Analytics fee (not for DS)
            if self.analytics == 'Y' and self.subject != "DS":
                self.total_cost += self.base_fee * 0.10
        else:
            print("Invalid subject")
            return

        # Hostel fee
        if self.hostel == 'Y':
            self.total_cost += 2000006

        # Food cost
        self.total_cost += self.food_months * 2000

        # Transportation cost
        if self.transport == "semester":
            self.total_cost += 13000
        elif self.transport == "annual":
            self.total_cost += 13000 * 2

        print("Total Annual Cost (INR):", self.total_cost)


# -------- User Input --------
try:
    subject = input("Enter subject (HR/Finance/Marketing/DS): ")
    analytics = input("Do you want Analytics (Y/N): ")
    hostel = input("Do you want Hostel (Y/N): ")
    food_months = int(input("Food required for how many months?: "))
    transport = input("Transportation (semester/annual): ")

    # Object Creation
    student1 = CourseFeeCalculator(subject, analytics, hostel, food_months, transport)

    # Method Call
    student1.calculate_fee()

except ValueError:
    print("Please enter valid numeric input for food months.")
```

46. Create a function to calculate age till now.

```python
from datetime import date

def calculate_age():
    year = int(input("Enter birth year: "))
    month = int(input("Enter birth month: "))
    day = int(input("Enter birth day: "))

    today = date.today()

    age = today.year - year

    # Adjust age if birthday hasn't come yet this year
    if (today.month, today.day) < (month, day):
        age -= 1

    print("Your age is:", age)

calculate_age()
```

47. Create a function to check age eligibility for given customer based on DOB. Function will take two input DOB and ELIGIBILITY age.

```python
def check_eligibility(birth_year, eligibility_age):
    today = date.today()
    current_age = today.year - birth_year

    if current_age >= eligibility_age:
        return "Eligible"
    else:
        return "Not Eligible"


year = int(input("Enter birth year: "))
eligibility = int(input("Enter eligibility age: "))

print(check_eligibility(year, eligibility))
```

51. Write a program to find largest number in the list.

```python
def find_largest(numbers):
    largest = numbers[0]

    for num in numbers:
        if num > largest:
            largest = num

    return largest



my_list = [10, 45, 23, 89, 67]

result = find_largest(my_list)
print("Largest number is:", result)
```

52. Write a program to check frequency of each element in the list.

```python
def check_frequency(lst):
    freq = {}

    for item in lst:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1

    return freq


my_list = [1, 2, 2, 3, 1, 4, 2, 3]

result = check_frequency(my_list)

for key, value in result.items():
    print(key, "occurs", value, "times")


```

53. There are two string l1 =[ 1,2,3,4,5] and l2 =[3,2,8,7,9] then write a program to find common elements in the list.

```python
def find_common(l1, l2):
    common = []

    for item in l1:
        if item in l2:
            common.append(item)

    return common


l1 = [1, 2, 3, 4, 5]
l2 = [3, 2, 8, 7, 9]

result = find_common(l1, l2)
print("Common elements:", result)
```
