Que 4)Create the function to calculate the maturity amount when deposit amount and interest rate is given. Interest rate can not be more than 7% and in negative.

```python
def cal_maturity(deposite, interest_rate):
    if interest_rate<0:
        return "rate cannot be negative"
    if interest_rate>7:
        return "rate cannot be more than 7%"

    interest = (deposite*interest_rate)/100
    maturity_amt = deposite + interest
    return maturity_amt


deposite = int(input("Enter deposite amount: "))
interest_rate = float(input("Enter interest rate in percent: "))

cal_maturity(deposite, interest_rate)
```

que 5)Create a function to calculate largest and smallest value from the list

```python
def cal_large_small(list):

    max = list[0]
    min = list[0]

    for i in list:
        if i > max:
            max = i

        if i < min:
            min = i

    return max, min

numbers = [10, 45, 2, 78, 23, 5]

max, min = cal_large_small(numbers)

print("Largest value:", max)
print("Smallest value:", min)
```

que 6) d1= {1:{'name': "nilesh", mobile: '789890678'},  2: {'name': "Jasmin", 'mobile': '6567788789'}   
create a function to acccept information from user like either id or name if user is entering name then roll number and mobile should be displayed. if id is given then name and mobile number should be displayed

```python
d1 = {1: {'name': "nilesh", 'mobile': '789890678'},
      2: {'name': "Jasmin", 'mobile': '6567788789'}}
def accept_info(data):
    x = input("Enter id or name: ")

    if x.isdigit():
        x = int(x)
        if x in data:
            print("Name is:", data[x]['name'])
            print("Mobile is:", data[x]['mobile'])
        else:
            print("Id not found")

    else:
        for key in data:
            if data[key]['name'] == x:
                print("Id is:", key)
                print("Mobile is:", data[key]['mobile'])
                break
        else:
            print("Name not found")


accept_info(d1)
```

Que 7) d1 = {'n':20, 'm':30, 'o':30}d2= {'n' : 47, 'x':20, "y":34}
d1 and d2 are two dictionaries. Merge two dictionaries if keys are different. for same keys add valuesoutput = {'n': 67, 'm':30, 'o':30, 'x':20, 'y':34}

```python
def merge_dict(d1, d2):
    result = d1.copy()

    for key, value in d2.items():
        if key in result:
            result[key] += value
        else:
            result[key] = value

    return result


d1 = {'n':20, 'm':30, 'o':30}
d2 = {'n':47, 'x':20, 'y':34}

merge_dict(d1, d2)
```
